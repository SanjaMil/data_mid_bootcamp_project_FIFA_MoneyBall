```python
#function for dropping columns with pictures and high nan value

def clean_data(dataset):
    for i in range(len(dataset.columns)):
        if dataset[dataset.columns[i]].isna().sum() >0.5*len(dataset):
            dataset = dataset.drop(dataset.columns[i], axis =1)
    return dataset
        
clean_data(data)
data.isna().sum()
```

```python
#dropping NANs and replacing NANs

data = data.drop(['LOAN_DATE_END', 'CLUB_LOGO', 'FLAG_PHOTO', 'PLAYER_PHOTO'], axis=1)
data = data.fillna(0)
```

```python
def convert_to_number(s):
    try:
        x = float(s)
    except:
        x = round(float(s.replace('€', '').replace('K','000').replace('M','000000')),0)
    return x
```

```python
data['WAGE'] = data['WAGE'].apply(convert_to_number)
data['VALUE'] = data['VALUE'].apply(convert_to_number)
data['RELEASE_CLAUSE'] = data['RELEASE_CLAUSE'].apply(convert_to_number)
data.dtypes
```

```python
CB_analysis = CB_num[['OVA','POT', 'AGE', 'VALUE', 'WAGE', 'TOTAL_STATS', 'BOV','HEADING_ACCURACY' , 'REACTIONS', 'INTERCEPTIONS', 'DEFENDING', 'MARKING', 'STANDING_TACKLE', 'DEF']]
CB_analysis
```

```python
#LinearRegression model - trying to predict OverallRating of the player
#splittinh numerical data set to target and all other
x = CB_analysis.drop(['TOTAL_STATS'], axis = 1)
y = CB_analysis['TOTAL_STATS']

#starting a model
lrm = LinearRegression()

#train the model
model = lrm.fit(x,y)
lrm.coef_
lrm.intercept_


```

```python
#evaluation of the model
sanja_model = lrm.predict(x)
#KPIs
mse = mean_squared_error(y, sanja_model)
rmse = mean_squared_error(y, sanja_model, squared = False)
mae = mean_absolute_error(y, sanja_model)
r2 = round(lrm.score(x,y), 2)


print("MSE = ", mse)

print("RMSE = ", rmse)

print("mae = ", mae)

print("R2 = ", r2)


```
